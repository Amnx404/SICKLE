```python
for key, values in final_epoch_values.items():
    if int(values) <= 0: continue
    all_pth = glob.glob(f'{key}/*tar')
    print(all_pth)
    for i in all_pth:
        if values in i.split('checkpoint')[1]:
            print(i)
            shutil.copy(i, i.split('checkpoint')[0]+'checkpoint_new.pth.tar')
        
```

```python
all_csv_files = glob.glob('./val_results/*.csv')
import pandas as pd


data_df = pd.DataFrame(columns=['Model','Dataset_Evaluation','Dataset_Model','Seed','MAPE'])
for i in all_csv_files:
    df = pd.read_csv(i)
    model_name = i.split('val_results/val_')[1].split('_direval')[0]
    dataset_eval = i.split('direval_')[1].split('_dirmodel')[0]
    dataset_model = i.split('dirmodel_')[1].split('_seed')[0]
    seed = i.split('seed')[1].split('.csv')[0]
    mape = df['val_mape'].values[0]
    data_df.loc[len(data_df)] = [model_name, dataset_eval, dataset_model, seed, mape]
    

    
```

```python
data_df.to_csv('./present/final_results.csv', index=False)
```
