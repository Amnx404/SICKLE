```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('Crop_Yield_Evaluation_Results_last_seed0.csv')

# List of unique models
models = df['Model'].unique()

# Determine the layout of the subplots
n_models = len(models)
ncols = 3  # Set the number of columns per row
nrows = (n_models + ncols - 1) // ncols  # Ensures enough rows to display all heatmaps

# Create a figure with subplots
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols * 5, nrows * 4))
axes = axes.flatten()  # Flatten the array to simplify loop indexing

# Find the global min and max MAPE values to set a common scale
global_min = df['MAPE'].min()
global_max = df['MAPE'].max()

# Loop through each model to create separate heatmaps
for i, model in enumerate(models):
    # Filter the dataframe for the current model
    model_data = df[df['Model'] == model]

    # Create a pivot table for the heatmap
    pivot_table = model_data.pivot_table(values='MAPE', index='Dataset_Evaluation', columns='Dataset_Model', aggfunc='mean')

    # Plot the heatmap in the corresponding subplot
    ax = axes[i]
    sns.heatmap(pivot_table, annot=True, cmap='BuPu', vmin=global_min, vmax=global_max, fmt=".2f", ax=ax)
    ax.set_title(f'Heatmap of MAPE for {model}')
    ax.set_xlabel('Dataset Model')
    ax.set_ylabel('Dataset Evaluation')

# Adjust the layout to prevent overlap
plt.tight_layout()

# Hide any unused axes if there are any
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.show()
```
